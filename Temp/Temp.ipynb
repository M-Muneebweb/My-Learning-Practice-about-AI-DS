```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('XglECD.csv')
```

```python
data['obtain_marks'] = data.iloc[:,2:].sum(axis = 1)
def grade(marks):
    if marks>80:
        return 'A'
    elif marks>70:
        return 'B'
    elif marks>50:
        return 'C'
    else:
        return 'F'
data['total_marks'] = 100
data['percentage'] = (data.obtain_marks / data.total_marks)*100
data["grades"] = data.percentage.apply(grade)
data['class'] = data['roll_no'].str.split('/').str[1]
```

```python
classes = data.groupby('class')
```

```python
for i in data['class'].unique():
    data2 = classes.get_group(i)
    x = data2.grades.value_counts().index
    y = data2.grades.value_counts().values
    plt.bar(x,y , color = ['red' , '#6af3f7' , '#5ffa2f' , '#f75cce'] , alpha = 1 , width = 1 )
    plt.title(f'{i} Grades Status')
    plt.ylabel('Counts')
    plt.xlabel('Grades')
    plt.show()
```
