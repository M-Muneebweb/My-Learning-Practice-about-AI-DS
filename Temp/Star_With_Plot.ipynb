```python
import numpy as np

std_mar = {
    "S101": {"English": 75, "Urdu": 82, "Math": 90, "Sindhi": 70, "Chemistry": 85},
    "S102": {"English": 65, "Urdu": 78, "Math": 88, "Sindhi": 60, "Chemistry": 80},
    "S103": {"English": 80, "Urdu": 85, "Math": 95, "Sindhi": 75, "Chemistry": 90},
    "S104": {"English": 70, "Urdu": 75, "Math": 85, "Sindhi": 68, "Chemistry": 78},
    "S105": {"English": 90, "Urdu": 88, "Math": 92, "Sindhi": 80, "Chemistry": 95}
}


highest  = {}
minimum  = {}
mean_val = {}

for sid, subjects in std_mar.items():
    marks = np.array(list(subjects.values()))
    highest[sid]  = np.max(marks)
    minimum[sid]  = np.min(marks)
    mean_val[sid] = np.mean(marks)


students   = np.array(list(mean_val.keys()))
mean_array = np.array(list(mean_val.values()))


rank_order = np.argsort(-mean_array)

ranks = {}
for pos, idx in enumerate(rank_order, start=1):
    ranks[students[idx]] = pos


for sid in std_mar:
    print(f"""Rank: {ranks[sid]}
    Student ID: {sid}    
    Highest Marks: {highest[sid]}
    Min Marks: {minimum[sid]}
{"*"*25}""")
```
