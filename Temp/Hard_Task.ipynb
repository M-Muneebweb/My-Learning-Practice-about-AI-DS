```python
def string(sen):
    for i in sen.split(' '):
        if len(i) >= 5:
            a = i[::-1]
            sen = sen.replace(i , a)
        else:
            continue
    return sen
```

```python
def checker(n):
    if n <= 1:
        return 'Not Prime'          # 1 ya negative prime nahi hote
    for i in range(2, n):
        if n % i == 0:         # agar kisi se divide hota hai
            return 'Not Prime'
    return 'Prime' 
```

```python
def repeat_counter(word):
    for i in word:
        word = word[1:]
        if i in word:
            print(f'{i} is first repeated word')
            break
        else:
            continue
    else:
        print('No Word is repeated')
        
```
