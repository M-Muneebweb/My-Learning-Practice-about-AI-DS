```python
import numpy as np
import math

my_temp_file = np.loadtxt("InvestmentDataset.csv", delimiter = ",",dtype = str)
head = my_temp_file[:1,:4]
my_file = np.delete(my_temp_file,0,axis = 0)

years = my_file[:,0].astype(float)
profit_temp = my_file[:,1].astype(str)
profit = np.array([float(p.replace('%', '')) for p in profit_temp])
add_inves = my_file[:,2].astype(float)
tol_amou = my_file[:,3].astype(float)



initial_inves = 5000
Total_inves = initial_inves + np.sum(add_inves)
Total_amount_in_bank = tol_amou[-1]
Arthimetic_mean = profit.mean()
Arthimetic_mean = round(Arthimetic_mean,2)
gmean = round((math.pow(np.prod((profit / 100) + 1), 1 / len(profit)) - 1) * 100,2)
profit_years = years[profit > 0]
loss_years = years[profit < 0]

print(f"""
                <<---- Investment Report ---->>

Initial Investment         :        {initial_inves}$
Total Investment           :        {Total_inves}$
Total Amount in Bank       :        {Total_amount_in_bank}$
Arithmetic Mean Profit     :        {Arthimetic_mean}%
Geometric Mean Profit      :        {gmean}%
Profit Making Years        :        {profit_years}
Loss Making Years          :        {loss_years}
""")
```
