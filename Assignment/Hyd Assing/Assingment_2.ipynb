```python
class DataBaseManagment:
    def add_data(self,Name,Email,Age,Gender):
        file = open("DataBase.txt","a+")
        file.write(f"{Name},{Email},{Age},{Gender}\n")
        file.seek(0)
        file.close()
    def del_data(self,email):
        ffile = open("DataBase.txt","r")
        d_list = []
        for line in ffile.readlines():
            if email != line.split(",")[1].strip():
                d_list.append(line)
                
            else:
                continue
        else:
            ffile.close()
            ffile = open("DataBase.txt","w")
            ffile.writelines(d_list)
            ffile.close()
    def __repr__(self,Email="fatima_01@yahoo.com"):
        file = open("DataBase.txt","r")
        for line1 in file.readlines():
            if Email == line1.split(",")[1].strip():
                return print(f"""
                Name   : {line1.split(",")[0].strip()}
                Email  : {line1.split(",")[1].strip()}
                Age    : {line1.split(",")[2].strip()}
                Gender : {line1.split(",")[3].strip()}""")
                break
        file.close()
    def all_users(self):
        da_list = []
        file = open("DataBase.txt","r")
        for line in file.readlines():
            da_list.append(line.replace("\n",""))
        print(da_list)
                

p1 = DataBaseManagment()


p1.add_data("Ali", "ali123@gmail.com", 17, "Male")
p1.add_data("Fatima", "fatima_01@yahoo.com", 19, "Female")
p1.add_data("Ahmed", "ahmed.khan@mail.com", 18, "Male")
p1.add_data("Zainab", "zainab_star@outlook.com", 20, "Female")
p1.add_data("Hassan", "hassan.dev@gmail.com", 21, "Male")
p1.add_data("Ayesha", "ayesha_786@hotmail.com", 22, "Female")
p1.add_data("Usman", "usman007@gmail.com", 23, "Male")
p1.add_data("Laiba", "laiba_love@gmail.com", 18, "Female")
p1.add_data("Saad", "saadcool@yahoo.com", 19, "Male")
p1.add_data("Mariam", "mariamxoxo@mail.com", 20, "Female")


#p1.del_data("hassan.dev@gmail.com")

#p1.all_users()

print(p1)


#p1.__repr__()
```
