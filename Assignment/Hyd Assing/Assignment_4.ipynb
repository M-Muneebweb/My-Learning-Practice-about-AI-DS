```python
import numpy as np

#dict
Student_data = {
    "st-1": {"Maths": 76, "Urdu": 91, "Sindhi": 58, "Chemistry": 87, "Physics": 69},
    "st-2": {"Maths": 93, "Urdu": 65, "Sindhi": 71, "Chemistry": 82, "Physics": 60},
    "st-3": {"Maths": 84, "Urdu": 74, "Sindhi": 97, "Chemistry": 69, "Physics": 99},
    "st-4": {"Maths": 59, "Urdu": 88, "Sindhi": 66, "Chemistry": 92, "Physics": 73},
    "st-5": {"Maths": 68, "Urdu": 79, "Sindhi": 54, "Chemistry": 90, "Physics": 81}
    }


#converting in numpy array



st_Id =    np.array([key for key in Student_data.keys()])
st_array = np.array([list (value.values()) for value in Student_data.values()])
subjects = np.array([list (key.keys()) for key in Student_data.values()])[0]

#creating ranks


lisst = []
for i in st_array:
    lisst.append(int(sum(i)))

ranks = []
for i in lisst:
    ranks.append(lisst.index(max(lisst)))
   
    lisst.insert(lisst.index(max(lisst)),0)
    lisst.remove(max(lisst))
ranks = np.array(ranks)
ranks = ranks+1

#Calculating subject with high score.
final = []
for i in list(ranks):
    high_sub = st_array[i-1]
    high_sub = high_sub.tolist()
    high_seq = subjects.tolist()
    final.append(high_seq[high_sub.index(max(high_sub))])

#Printing result.
count = 1
for i in list(ranks):
    print(f"""
    ***************
    Rank: {count}
    Student Id: st-{i}
    Subject with high score: {final[count-1]}
    Score: {count}
    ***************""")
    count += 1
```
