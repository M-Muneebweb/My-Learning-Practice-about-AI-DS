```python
import numpy as np

product_f = np.loadtxt("product.csv",delimiter=",",dtype=str)
headp = product_f[:1]
product_f = np.delete(product_f,0,axis=0) 
order_f = np.loadtxt("order.csv",delimiter=",",dtype=str)
heado = order_f[:1]
order_f = np.delete(order_f,0,axis=0) 
customer_f = np.loadtxt("customer.csv",delimiter=",",dtype=str)
headc = customer_f[:1]
customer_f = np.delete(customer_f,0,axis=0) 

cusid_ord = order_f[ : , 1 ]
cusid_cus = customer_f[ : , 0]
proid_ord = order_f[ : , 2]
price_ord = order_f[ : , 4].astype(float)
quant_ord = order_f[ : , 3].astype(int)



#q1 repeating customers

unique_ids , counts = np.unique(cusid_ord, return_counts=True)
repeating_customers = unique_ids[counts > 1]

#q2,3 top and least selling products
p_uni_id , counts1 = np.unique(proid_ord , return_counts=True)
max_count_pro = np.max(counts1)
top_sell_pro = p_uni_id[counts1 == max_count_pro]
min_count_pro = np.min(counts1)
least_sell_pro = p_uni_id[counts1 == min_count_pro]

#q5 more than 5 times visitors

more_10_vis = unique_ids[counts > 10]
more_5_vis = unique_ids[counts > 5]

#q6 total income

tot_inc = np.sum(price_ord * quant_ord)


#printing all data
repeating_customers = repeating_customers.reshape(10,2)
more_5_vis = more_5_vis.reshape(10,2)


print(f"""
  Repeated Customers
-----------------------
{repeating_customers} 

  Top Saled Products
 ---------------------
 {top_sell_pro}
 
 Least Saled Products
-----------------------
 {least_sell_pro}

 More than 5 times visitors
-----------------------------
{more_5_vis}

 Total income
--------------
  {tot_inc}
""")
```
