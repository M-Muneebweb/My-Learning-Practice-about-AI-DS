```python
import numpy as np
import math
invesment = 2500
profit = np.array([0.25,0.5,1,-0.5,0.25])
first = ((invesment/100)*25)+invesment
second = ((first/100)*50)+first
third = ((second/100)*100)+second
fourth = ((third/100)*-50)+third
fifth = ((fourth/100)*25)+fourth
factor = profit + 1
product = np.prod(factor)
gmean = math.pow(product,1/len(profit))-1
print(f"""---- Your invesment is: {invesment}
---- Arthimetic mean: {round(np.mean(profit)*100,2)}%
---- Geometric mean: {round(gmean*100,2)}%
---- Total amount in bank: {fifth}$""")
```

```python
print(first)
print(second)
print(third)
print(fourth)
print(fifth)
```
