```python
import math
point1 = []
point1.append(int(input('Enter first  point x-value: ')))
point1.append(int(input('Enter first  point y-value: ')))
point2 = []
point2.append(int(input('Enter second point x-value: ')))
point2.append(int(input('Enter second point y-value: ')))
def dis_finder(p1 , p2):
    ans = math.sqrt(((p2[0]-p1[0])**2)+(p2[1]-p1[1])**2)
    return ans
print(dis_finder(p1 = point1,p2 = point2))
```
