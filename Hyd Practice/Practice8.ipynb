```python

import matplotlib.pyplot as plt
```

```python
x = [80 , 110 , 120 , 150 , 200]
y = [30 , 50 , 60 , 90 , 100]
plt.plot(x,y)
plt.title("Graph heading name")
plt.xlabel("X-axis name")
plt.ylabel("Y-axis name")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x,y1 , label = 'y = x**2' , color = 'green')
plt.plot(x,y2 , label = 'y = x**3' , color = 'red')
plt.title("Graph heading name")
plt.xlabel("X-axis name")
plt.ylabel("Y-axis name")
plt.grid(True)
plt.legend()
plt.savefig('practice_graph.png')
plt.show()
```
