```python
import numpy as np
import pandas as pd
```

```python


dice1 = np.random.choice([1,2,3,4,5,6], size = 100)
dice2 = np.random.choice([1,2,3,4,5,6], size = 100)


count = 0
for one,two in zip(dice1,dice2):
    if one+two == 7:
        count+=1
    else:
        continue

print(f"Emperical Probability is: {count/100}")
```
