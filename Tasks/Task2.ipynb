```python
import numpy as np
import pandas as pd
ss = ['B']*5 + ['R']*3 + ['G']*2

print(f'Theoritical Probability of :  {ss.count('B')/len(ss)}')
```

```python
ss1 = np.random.choice(ss , size = 10000)



ss1 = pd.Series(ss1)


for i in ss1:
    
    print(f'Theoritical Probability of :  {ss1.value_counts(normalize = True)[i]}')
```
